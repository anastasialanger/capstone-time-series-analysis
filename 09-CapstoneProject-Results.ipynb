## 9. Summary of results

<li> The goal of this project was to investigate whether it is possible to achieve good forecast results for a one time-series using a machine learning model with the set of hyperparameters tuned on a different time-series. The shape of both time-series should be similar. </li> <li> To identify the time-series of a similar shape, a clustering algorithm was applied to a subset of time-series from the selection of electricity demand time-series. </li> <li> The time-series of 20 randomly selected meters were clustered into 3 groups using hierarchical clustering and DTW distances computed between the pairwise combinations of time-series. </li> <li> Using a one time-series from each cluster, the parameters of machine learning models were tuned. Afterwards these parameters were used to compute the forecast for a different time-series from the same cluster. </li> <li> Table below summarizes prediction MAE obtained with each machine learning model for a time-series from each cluster. </li>

```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
results = \
pd.DataFrame({'Model': ['Avg-baseline', 'kNN','Random Forest','2L Neural Network','ConvNet'], 'Pred MAE Cluster 1': [0.13,0.04,0.06,0.03,0.05],
             'Pred MAE Cluster 2': [0.11,0.04,0.05,0.04,0.04], 'Pred MAE Cluster 3': [0.08,0.04,0.04,0.05,0.07]})
results
```

```python
for i in results.columns[1:]:
    pred_mae = results[i]
    models = results['Model']
    
    xcor = np.arange(len(pred_mae))
    plt.bar(xcor, pred_mae)
    plt.xticks(xcor, models, rotation = 20)
    plt.title(i)
    plt.show()
```

Plot forecasts obtained with each model for each cluster:

```python
baseline_meter481 = pd.read_pickle('Baseline_meter481')
kNN_meter481 = pd.read_pickle('kNN_meter481')
RandForest_meter481 = pd.read_pickle('RandForest_meter481')
NeuralNetwork_meter481 = pd.read_pickle('NeuralNetwork_meter481')
ConvNet_meter481 = pd.read_pickle('ConvNet_meter481')
```

```python
plt.figure(figsize=(20,10))
plt.plot(kNN_meter481[kNN_meter481['type']=='test'].date, \
         kNN_meter481[kNN_meter481['type']=='test'].pred_value, linewidth=1, color='black', label='kNN')
plt.plot(RandForest_meter481[RandForest_meter481['type']=='test'].date, \
         RandForest_meter481[RandForest_meter481['type']=='test'].pred_value, linewidth=1, color='blue', label='RandForest')
plt.plot(baseline_meter481[baseline_meter481['type']=='test'].date, \
         baseline_meter481[baseline_meter481['type']=='test'].pred_value, linewidth=2, color='red', label='Baseline')
plt.legend(loc='upper right', ncol=1)
plt.xlabel('Date')
plt.ylabel('Forecasted normalized demand')
plt.title('Cluster 1')
plt.show();

plt.figure(figsize=(20,10))
plt.plot(NeuralNetwork_meter481[NeuralNetwork_meter481['type']=='test'].date, \
         NeuralNetwork_meter481[NeuralNetwork_meter481['type']=='test'].pred_value, linewidth=1, color='orange', label='2L NeuralNet')
plt.plot(ConvNet_meter481[ConvNet_meter481['type']=='test'].date, \
         ConvNet_meter481[ConvNet_meter481['type']=='test'].pred_value, linewidth=1, color='green', label='ConvNet')
plt.legend(loc='upper right', ncol=1)
plt.xlabel('Date')
plt.ylabel('Forecasted normalized demand')
plt.title('Cluster 1')
plt.show();
```

<li> Machine learning models always perform better in terms of prediction MAE than the baseline model computed as an average of historical demand. </li> <li> Using a 24 hours lag in demand as an additional feature improved the performance of all machine learning models. </li> <li> None of machine learning models achieves consistently lower MAE across all clusters. </li> <li> Parameters of machine learning models tuned using the data from a one time-series from a cluster can be used to configure a machine learning model and compute the forecast for a different time-series from the same cluster without a significant degradation of forecast's quality. When computing the forecasts for many meters, it can significantly reduce calculation time because the parameters can be tuned only on one sample.</li>
